```python
#Create data vehicles
def createData():
    vehicles=[[1,5000],
         [2,5000],
         [3,5000],
         [4,5000],
         [5,5000]]
    vehiclesDataFrame=pd.DataFrame(np.matrix(vehicles))
    vehiclesDataFrame.head()
    clientesDataFrame=pd.read_csv("clientes.csv")
    clientesDataFrame.dropna(axis=0, how='any', subset=['Lat','Lon'], inplace=True)
    #X=clientesDataFrame.loc[:,['NameClient','Lat','Lon']]
    return vehiclesDataFrame,clientesDataFrame
#Asignación variables de latitud y longitud.

def getModelKmeans(cdf,km):
    return [km[i].fit(cdf[['Lat']]).score(cdf[['Lat']]) for i in range(len(km))]


def createKmeans(cdf):
    X=cdf.loc[:,['NameClient','Lat','Lon']]
    kmeans = [KMeans(n_clusters=k) for k in range(1,10)]
    performance=getModelKmeans(cdf,kmeans)
    
def calculateKm(cdf,vdf):
    nclus= math.ceil(math.ceil(cdf['Carga'].sum()/vdf[1].mean())/2)
    cdf['cln']= KMeans(n_clusters=nclus).fit_predict(cdf[['Lon', 'Lat']])

def processCluster(cdf):
    data=[]
    cargatotal=[]
    for j in range(8):
        data.append(cdf[(cdf['cln'] == j)])
        cargatotal.append(data[j]['Carga'].sum())
    print(cargatotal)

    
def normalization(dataX):
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(dataX)
    X_normalizada = normalize(X_scaled)
    X_normalizada = pd.DataFrame(X_normalizada)
    X_normalizada.columns = X.columns 

def includeLabel(dataX):
    X_ = dataX 
    X_['ckO'] = labels
    dn = X1
    dn.head()
    
def processKmeansOpt(cdf):
    data0 = cdf[(cdf['cln'] == 0)]
    data1 = cdf[(cdf['cln'] == 1)]
    data2 = cdf[(cdf['cln'] == 2)]
    data3 = cdf[(cdf['cln'] == 3)]
    data4 = cdf[(cdf['cln'] == 4)]
    data5 = cdf[(cdf['cln'] == 5)]
    data6 = cdf[(cdf['cln'] == 6)]
    data7 = cdf[(cdf['cln'] == 7)]
    d0 = data0
    d1 = data1 
    d2 = data2 
    d3 = data3 
    d4 = data4 
    d5 = data5 
    d6 = data6 
    d7 = data7
    cargat0 = data0['Carga'].sum()
    cargat1 = data1['Carga'].sum()
    cargat2 = data2['Carga'].sum()
    cargat3 = data3['Carga'].sum()
    cargat4 = data4['Carga'].sum()
    cargat5 = data5['Carga'].sum()
    cargat6 = data6['Carga'].sum()
    cargat7 = data7['Carga'].sum()
    concat = [cdf[(cdf['NameClient'] == 'Depot')], cdf[(cdf['cln'] == 0)]]
    concat0 = pd.concat(concat, sort='False', ignore_index='True')
    concat0.drop_duplicates('NameClient', inplace=True) #Eliminar duplicado del depot en caso de que el 
                                                          #cluster incluya el depósito.
    data0 = concat0
    drop_features = ['Lat','Lon','cln','ID','NameClient','Ciudad'] 
    X = data0.drop(drop_features, axis = 1) 

    #Filtro de los valores perdidos (si los hay)
    X.fillna(method ='ffill', inplace = False)
    scaler = StandardScaler() 
    X_scaled = scaler.fit_transform(X) 

    #Normalizar los datos para que los datos sigan aproximadamente una distribución gaussiana.
    X_normalizada = normalize(X_scaled) 

    #Conversión de la matriz numpy en un DataFrame de pandas
    X_normalizada = pd.DataFrame(X_normalizada) 

    #Renombre de las columnas
    X_normalizada.columns = X.columns 
    optics_model = OPTICS(min_samples = 11, xi = 0.06, min_cluster_size = 0.08) 
    optics_model.fit(X_normalizada) 
    labels = optics_model.labels_[optics_model.ordering_]
    datadX = data0 
    datadX['lab0'] = labels
    dn = datadX
    data1n = dn[(dn['lab0'] == -1)]
    data0 = dn[(dn['lab0'] == 0)]
    d1n = data1n 
    d0 = data0
    Qt1n = data1n['Carga'].sum()
    Qt0 = data0['Carga'].sum()
    Qt1 = data1['Carga'].sum()
    QR0 = Qt1n
    QR1 = Qt0
    QR2 = Qt1
    datak0 = dn
    return datak0
    
def semiverseno(cdf):
    cdf['Lat'] = np.radians(cdf['Lat'])
    cdf['Lon'] = np.radians(cdf['Lon'])
    dist = DistanceMetric.get_metric('haversine')
    EM = dist.pairwise(cdf[['Lat','Lon']].to_numpy()) * 6371.01 #Radio de la tierra en km.
    ExportMatriz = pd.DataFrame(EM, columns=cdf.NameClient.unique(), index=cdf.NameClient.unique())
    #print(ExportMatriz)
    
def clkmeans(datak0):
    DtF = datak0[(datak0['lab0'] == -1)]
    dataDepot = datak0[(datak0['NameClient'] == 'Depot')]
    concat = [dataDepot, DtF]
    concatt = pd.concat(concat, sort='False', ignore_index='True')
    concatt.drop_duplicates('NameClient', inplace=True) #Eliminar duplicado del depot en caso de que el 
    DF = concatt
    return DF
def drawPath(DF):
    NameClient = DF.loc[:, 'NameClient']
    IDMatriz = np.array(NameClient)
    nodos = IDMatriz
    nodos = len(nodos)
    n = nodos
    clientes = [i for i in range(n)]
    arcos = [(i,j) for i in range (n) for j in range (n) if i!=j]
    Longitud = DF.loc[:, 'Lon']
    Lonmatriz = np.array(Longitud)
    x = Lonmatriz
    Latitud = DF.loc[:, 'Lat']
    Latmatriz = np.array(Latitud)
    y = Latmatriz
    x = Lonmatriz
    y = Latmatriz
    distancia = {(i,j):np.hypot(x[i]-x[j], y[i]-y[j])*111.138746329478 for i,j in arcos}
    #Solución inicial con Nearest Neighbor.
#Para iniciar una solución con local search 2 opt, se necesita de una solución previa.
    starting_node = 0
    NN = Nearest_neighbor(starting_node, clientes, distancia)

    #Distancia total de la solución Nearest Neighbor.
    d = total_distancia(NN, distancia)

    #Extraer depósito.
    depotmatriz = np.column_stack((Lonmatriz, Latmatriz))
    depot = depotmatriz[:1]

    #Latitud y longitud del depósito.
    depotlon = depot[:,0]
    depotlat = depot[:,1]
    time_inicio = time.time()
    sol = NN.copy()

    cambio = 1
    count = 0
    
    while cambio != 0:
        count = count + 1
        inicial = total_distancia(sol, distancia)
        sol = LS_2opt(sol, distancia).copy()
        final = total_distancia(sol, distancia)

        cambio = np.abs(final - inicial)

    time_final = time.time()
    #Gráfica de la solución local search.
    plt.rcParams['figure.figsize'] = [8, 6]

    #Nombre de los nodos.
    #for n in range(len(x)):
     #   plt.annotate(str(n), xy=(x[n], y[n]), xytext=(x[n], y[n]), color='black')

    #Graficar ruta de local search.
    for n in range(len(sol)-1):
        plt.plot([x[sol[n]], x[sol[n+1]]], [y[sol[n]], y[sol[n+1]]], color='blue')

    mapa = plt.plot(x, y, marker='.', linewidth=0, color='#e69500')
    mapa = plt.plot(depotlon, depotlat, marker='x', linewidth=0, color='black')
    mapa = plt.title('Gráfica de la ruta con nuevo punto de localización', family='Arial', fontsize=14)
    mapa = plt.xlabel('Coordenada x = Longitud')
    mapa = plt.ylabel('Coordenada y = Latitud')
    mapa = plt.show()



    
def Nearest_neighbor(starting_node, clientes, distancia):
    NN = [starting_node]
    n = len(clientes)
    
    while len(NN) < n:
        k = NN[-1]
        nn = {(k,j): distancia[(k,j)] for j in clientes if j!=k and j not in NN}
        nn.items()
        new = min(nn.items(), key = lambda x: x[1])
        NN.append(new[0][1])
    
    NN.append(starting_node)
    
    return NN


def total_distancia(lista, distancia):
    
    dist = 0
    for n in range(len(lista)-1):
        i = lista[n]
        j = lista[n+1]
        dist = dist + distancia[(i,j)]
    
    return dist


def LS_2opt(NN, distancia):
    
    min_change = 0
    
    for i in range(len(NN)-2):
        for j in range(i+2, len(NN)-1):
            
            costo_actual = distancia[(NN[i], NN[i+1])] + distancia[(NN[j], NN[j+1])]
            costo_nuevo = distancia[(NN[i], NN[j])] + distancia[(NN[i+1], NN[j+1])]
            change = costo_nuevo - costo_actual
            
            if change < min_change:
                min_change = change
                min_i = i
                min_j = j
                
    if min_change < 0:
        NN[min_i+1: min_j+1] = NN[min_i+1: min_j+1][::-1]
        
    return NN

#Clientes == puntos.
#Depósito == x.
    
vdf,cdf=createData()
createKmeans(cdf)
calculateKm(cdf,vdf)
processCluster(cdf)
dk0=processKmeansOpt(cdf)
semiverseno(cdf)

drawPath(clkmeans(dk0))
```
